```python
n1 = int(input("Digite um numero: "))

if n1 >= 0:
    if n1 % 2 == 0:
        print("\n=====| É multiplo de 2! |=====\n")
        
        if n1 % 3 == 0:
            print("\n=====| Também é multiplo de 3! |=====\n")
        else:
            print("\n=====| Não é multiplo de 3! |=====\n")
    else:
        print("\n=====| Não é multiplo de 2! |=====\n")
else:
    print("\n=====| O numero é negativo! |=====\n")
```
