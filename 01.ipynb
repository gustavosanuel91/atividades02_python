```python
idade = int(input("Digite sua idade: "))
ingresso = input("Voçê tem ingresso?(s/n): ").lower().strip()

if idade >= 12:
    if ingresso == "s":
        print("\n=====| Acesso Liberado! |=====\n")
    
if idade >= 12:
    if ingresso == "n":
        print("\n=====| Acesso negado, ingresso necessario! |=====\n")
    
else:
    print("\n=====| Acesso negado, idade necessaria! |=====\n")
```
