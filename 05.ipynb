```python
idade = int(input("Digite sua idade: "))
documento = input("Voçê tem documento?(s/n: ").lower()

if idade >= 16:
    if documento == "s":
        print("\n=====| PODE ENTRAR! |=====\n")
    else:
        print("\n=====| PRECISA APRESENTAR DOCUMENTO! |=====\n")
else:
    print("\n=====| PRECISA TER MAIS DE 16 ANOS! |=====\n")
```
