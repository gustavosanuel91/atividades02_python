```python
valor = int(input("Digite o valor da compra: "))
cupom = input("Voçê tem cupom? (s/n): ").lower()

if valor >= 100:
    if cupom == "s":
        print("\n=====| DESCONTO APLICADO! |=====\n")
    else:
        print("\n=====| SEM DESCONTO |=====\n")
else:
    print("\n=====| VALOR INSUFICIENTE PARA DESCONTO! |=====\n")
```
