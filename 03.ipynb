```python
nota = int(input("Digite sua nota: "))
frequencia = int(input("Digite a frequencia: "))

if nota >= 7:
    if frequencia >= 75:
        print("\n=====| Aprovado! |=====\n")
    else:
        print("\n=====| Reprovado por falta! |=====\n")

else:
    print("\n=====| Reprovado por nota! |=====\n")
```
