```python
nome = input("Digite seu nome: ").strip()
idade = int(input("Digite sua idade: "))

if nome:
    if idade >= 18:
        print("\n=====| Cadastro completo! |=====\n")
    else:
        print("\n=====| Cadastro incompleto, menor de idade! |=====\n")
else:
    print("\n=====| Nome obrigatorio! |=====\n")
```
