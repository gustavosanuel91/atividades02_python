```python
idade = int(input("Digite sua idade: "))
dor = input("Sente dor forte? (s/n): ").lower()

if idade >= 60:
    if dor == "s":
        print("\n=====| Atendimento prioritario imediato! |=====\n")
    else:
        print("\n=====| Aguardando triagem |=====\n")
else:
    print("\n=====| Atendimento comum! |=====\n")
```
